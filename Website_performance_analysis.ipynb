```python

import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_web = pd.read_csv(r"A:\Website Analysis\data-export (1).csv")
```

```python
df_web.head()
```

```python
df_web.columns =  df_web.iloc[0]  # this removed  unnamed column in data set 
df_web.head()
```

```python
df_web.columns = df_web.iloc[0]  #change the indexing value with iloc function in all columns
df_web = df_web.drop(index = 0).reset_index(drop = True)  # drop the 0 indexing duplicate values 
df_web.columns = ["channal group", "DateHours", "Users", "Sessions", "Engaged Sessions", "Average engagement time per session", "Engaged sessions per user", "Events per session",	"Engagement rate",	"Event count"]

# using columns to change columns names for all columns values
```

```python
df_web.head()
```

```python
df_web["DateHours"] = pd.to_datetime(df_web["DateHours"], format="%Y%m%d%H", errors="coerce")
#  change the formate of date time objective formate to date time formate using datetime function                                    
```

```python
df_web.head()
```

```python
df_web.info()
```

```python
numeric_cols = df_web.columns.drop(["channal group", "DateHours"]) # its meand  These two columns are non-numeric, 
#  and you want to apply numeric processing only to the remaining columns.
# So numeric_cols will contain all column names that are expected to be numeric (like "Users", "Sessions", etc.)



df_web[numeric_cols] = df_web[numeric_cols].apply(pd.to_numeric, errors="coerce")
# For all columns in numeric_cols, convert their values to proper numeric types (int or float).
# If any value is invalid (like a string or symbol in a numeric column), it will be replaced with NaN due to errors="coerce".

df_web["Hour"] = df_web["DateHours"].dt.hour
# Extracts the hour part (0 to 23) from the datetime column "DateHours".
# Stores that hour value in a new column called "Hour".

```

# Que 4:  How Does Engagement rate vary across different traffice Channal?

# Engagement Rate Destribution by channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_web, x="channal group", y="Engagement rate", hue="channal group", palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Que 5: Which Channals are driving more engaged sessions compared to non-engaged ones, and what strategies can improve engagement in underperforming channelsd?

```python
 df_web.head()
```

# Engaged vs Non-Engaged Sessions

```python
session_df_web = df_web.groupby("channal group")[["Sessions", "Engaged Sessions"]].sum().reset_index()
session_df_web["Non-Engaged"] = session_df_web["Sessions"] - session_df_web["Engaged Sessions"]
session_df_web_melted = session_df_web.melt(id_vars="channal group", value_vars=["Engaged Sessions", "Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_web_melted, x="channal group", y="value", hue="variable")
plt.title(" Engaged va Non-Engaged Session")
plt.xticks(rotation=45)
plt.show()
```

# Que 6: At what hours of the day does each channal drive the most traffice?

```python
df_web.head()
```

# Traffice by hours and channnal

```python
heatmap_data = df_web.groupby(["Hour", "channal group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(13,7))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffice by Hours and Channal")
plt.xlabel("channal Group")
plt.ylabel("Hours of Day")
plt.show()
```

# Que 7:  is there any correlation between high traffice (Session) and high engagement rate over time?

# Engagement rate Distribution by Channal

```python
df_plot = df_web.groupby("DateHours")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHours"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DataHours"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement rate  vs Sessions over time")
plt.xlabel("DataHours")
plt.legend()
plt.grid(True)
plt.show()
```
